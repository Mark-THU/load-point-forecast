# classical forecast methods

```python
import random
import numpy as np
import pandas as pd
import pdb
import matplotlib.pyplot as plt

from tqdm import tqdm
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

## data prepare

### data load

```python
url = '../data/beijing.csv'
data = pd.read_csv(url, index_col='time')
load = data.load.values
```

### build supervised data

```python
def series_to_supervise(data, seq_len, target_len):
    dim_0 = data.shape[0] - seq_len - target_len + 1
    dim_1 = data.shape[1]
    x = np.zeros((dim_0, seq_len, dim_1))
    y = np.zeros((dim_0, target_len, dim_1))
    for i in range(dim_0):
        x[i] = data[i:i + seq_len]
        y[i] = data[i + seq_len:i + seq_len + target_len]
    print("supervised data: shape of x: {}, shape of y: {}".format(
        x.shape, y.shape))
    return x, y
```

## SARIMA

### test model

```python
def test_sarima(test_data, seq_len, target_len, config):
    p = config['p']
    d = config['d']
    q = config['q']
    P = config['P']
    D = config['D']
    Q = config['Q']
    S = config['S']
    inputs, labels = series_to_supervise(test_data, seq_len, target_len)
    inputs = inputs[:, :, -1]
    labels = labels[:, :, -1]
    pred = []
    true = []
    loop = tqdm(range(len(inputs)))
    for i in loop:
        model = SARIMAX(inputs[i],
                        order=(p, d, q),
                        seasonal_order=(P, D, Q, S))
        model_fit = model.fit(disp=False)
        pred.append(model_fit.predict(seq_len, seq_len + target_len - 1))
        true.append(labels[i])
        loop.set_description('Epoch:{}/{}'.format(i + 1, len(inputs)))
    load_pred = np.array(pred)
    load_true = np.array(true)
    MAPE = np.mean(np.abs(load_true - load_pred) / load_true)
    MAE = np.mean(np.abs(load_true - load_pred))
    RMSE = np.sqrt(np.mean(np.square(load_true - load_pred)))
    return MAPE, MAE, RMSE, load_pred, load_true
```

### hyper-parameters config

```python
test_data = data[-500:].values
seq_len = 240
target_len = 24
config = {'p': 1, 'd': 1, 'q': 1, 'P': 1, 'D': 1, 'Q': 1, 'S': 24}
MAPE, MAE, RMSE, load_pred, load_true = test_sarima(test_data, seq_len, target_len, config)
```

```python
print('MAPE:{}, MAE:{}, RMSE:{}'.format(MAPE, MAE, RMSE))
```

### figure plot

```python
plt.figure(figsize=(15, 7))
load_pred = load_pred.reshape(-1, target_len)
load_true = load_true.reshape(-1, target_len)
plt.plot(load_pred[:200,23], 'r')
plt.plot(load_true[:200,23], 'g')
plt.ylim(0)
```

## SARIMAX

### test model

```python
def test_sarimax(test_data, seq_len, target_len, config):
    p = config['p']
    d = config['d']
    q = config['q']
    P = config['P']
    D = config['D']
    Q = config['Q']
    S = config['S']
    inputs, labels = series_to_supervise(test_data, seq_len, target_len)
    inputs_exog = inputs[:, :, 0]
    inputs = inputs[:, :, -1]
    labels_exog = labels[:, :, 0]
    labels = labels[:, :, -1]
    pred = []
    true = []
    loop = tqdm(range(len(inputs)))
    for i in loop:
        model = SARIMAX(inputs[i],
                        exog=inputs_exog[i],
                        order=(p, d, q),
                        seasonal_order=(P, D, Q, S))
        model_fit = model.fit(disp=False)
        pred.append(model_fit.predict(seq_len, seq_len + target_len - 1, exog=labels_exog[i]))
        true.append(labels[i])
        loop.set_description('Epoch:{}/{}'.format(i+1, len(inputs)))
    load_pred = np.array(pred)
    load_true = np.array(true)
    MAPE = np.mean(np.abs(load_true - load_pred) / load_true)
    MAE = np.mean(np.abs(load_true - load_pred))
    RMSE = np.sqrt(np.mean(np.square(load_true - load_pred)))
    return MAPE, MAE, RMSE, load_pred, load_true
```

### hyper-parameters config

```python
test_data = data[-500:].values
seq_len = 240
target_len = 24
config = {'p': 1, 'd': 1, 'q': 1, 'P': 1, 'D': 1, 'Q': 1, 'S': 24}
MAPE, MAE, RMSE, load_pred, load_true = test_sarimax(test_data, seq_len, target_len, config)
```

```python
print('MAPE:{}, MAE:{}, RMSE:{}'.format(MAPE, MAE, RMSE))
```

### figure plot
